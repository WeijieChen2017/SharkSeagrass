```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_x_y(x, y):
    fig, ax = plt.subplots(figsize=(5, 5), dpi=100)
    ax.scatter(x, y, s=3)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.xaxis.tick_bottom()
    ax.yaxis.tick_left()
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    fig.tight_layout()
    fig.show()
```

```python
def plot_x_yList(x, yList):
    fig, ax = plt.subplots(figsize=(5, 5), dpi=100)
    for y in yList:
        ax.scatter(x, y, s=3)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.xaxis.tick_bottom()
    ax.yaxis.tick_left()
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    fig.tight_layout()
    fig.show()
```

```python
def ReLU(x, y):
    y[y<0] = 0
    return y
```

```python
# x, wx+b, sigma(wx+b), sum(sigma(wx+b))

w = 1
b = -0.5

n_points = 2000
xmesh = np.asarray(range(1, n_points)) / 1000 - 1
y1 = w*xmesh+b
y1 = ReLU(xmesh, y1)

plot_x_y(xmesh, y1)



```

```python
# x, wx+b, sigma(wx+b), sum(sigma(wx+b))
param_pairs = [
    [1, -0.5],
    # [1, 0.5],
    [1, -0.25],
    # [1, 0.25]
]


n_points = 2000
xmesh = np.asarray(range(1, n_points)) / 1000 - 1
yList = []
for w, b in param_pairs:
    y = w*xmesh+b
    y = ReLU(xmesh, y)
    yList.append(y)

plot_x_yList(xmesh, yList)



```

```python
# x, wx+b, sigma(wx+b), sum(sigma(wx+b))
param_pairs = [
    [[1, -0.25],[1, -0.5]],
    [[1, -0.5], [1, -0.75]],
]


n_points = 2000
xmesh = np.asarray(range(1, n_points)) / 1000 - 1
yList = []
for p1, p2 in param_pairs:
    w1, b1 = p1
    w2, b2 = p2
    y1 = w1*xmesh+b1
    y1 = ReLU(xmesh, y1)
    y2 = w2*xmesh+b2
    y2 = ReLU(xmesh, y2)
    yList.append(y1-y2)


plot_x_yList(xmesh, yList)



```

```python
# x, wx+b, sigma(wx+b), sum(sigma(wx+b))
param_pairs = [
    [[[1, -0.25],[1, -0.5]], [[1, -0.5], [1, -0.75]]],
]


n_points = 2000
xmesh = np.asarray(range(1, n_points)) / 1000 - 1
yList = []
for s1, s2 in param_pairs:
    p1, p2 = s1
    w1, b1 = p1
    w2, b2 = p2
    y1 = w1*xmesh+b1
    y1 = ReLU(xmesh, y1)
    y2 = w2*xmesh+b2
    y2 = ReLU(xmesh, y2)
    z1 = y1-y2

    p1, p2 = s2
    w1, b1 = p1
    w2, b2 = p2
    y1 = w1*xmesh+b1
    y1 = ReLU(xmesh, y1)
    y2 = w2*xmesh+b2
    y2 = ReLU(xmesh, y2)
    z2 = y1-y2
    
    yList.append(z1-z2)


plot_x_yList(xmesh, yList)



```

```python
filepath = "x_noisy_76.npy"

import numpy as np

data = np.load(filepath)
print(data.shape)

import matplotlib.pyplot as plt

# plot 8 images
fig = plt.subplots(2, 8, figsize=(16, 2))
for i in range(8):
    img = data[i]
    plt.subplot(2, 8, i+1)
    plt.hist(img.flatten(), bins=100)

    plt.subplot(2, 8, i+9)
    img = img / 3 # -1 to 1
    img = img + 1
    img = img / 2
    # change the img from c, w, h to w, h, c
    img = np.transpose(img, (1, 2, 0))
    img = np.rot90(img, 1)
    plt.imshow(img)

plt.show()
```
